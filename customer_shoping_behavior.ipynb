```python
import pandas as pd
df =  pd.read_csv("customer_shopping_behavior.csv")
df.head()
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
df.isnull().sum()
```

```python

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')

df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}) 
print(df.columns)
```

```python
# create a new column age_group
labels = ['young adult', 'adult','middle aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4 ,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequenct_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90

}
df['frequency_of_purchases_days'] = df['frequency_of_purchases'].map(frequenct_mapping)
print(df[['frequency_of_purchases','frequency_of_purchases_days']].head(10))
print("\n")
print(df['frequency_of_purchases_days'].isnull().sum())
```

```python
(df[['discount_applied','promo_code_used']]).head()
(df['discount_applied'] == df['promo_code_used']).all()
df = df.drop('promo_code_used',axis=1)
```

```python
import sys
import subprocess

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

try:
    from sqlalchemy import create_engine
    import pymysql
    print("SQLAlchemy and PyMySQL are successfully imported!")
except ImportError:
    print("Modules not found. Installing sqlalchemy and pymysql...")
    install('sqlalchemy')
    install('pymysql')
    from sqlalchemy import create_engine
    import pymysql
    print("Successfully installed and imported modules!")

engine = create_engine('mysql+pymysql://root:@localhost/customer')
try:
    df.to_sql('customers', con=engine, if_exists='replace', index=False)
    print(f"Successfully sent {len(df)} rows to XAMPP!")
except Exception as e:
    print(f"Error: {e}")
    print("\nTIP: Make sure XAMPP is running and the database 'customer' exists in phpMyAdmin.")
```
